```python
!pip install torch torchvision --quiet

import os
import random
import numpy as np
import pandas as pd
from PIL import Image

import torch
import torchvision.transforms as transforms
from torchvision import models

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
```

```python
# Cargar ResNet18 preentrenada

model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# Quitamos la última capa de clasificación
model = torch.nn.Sequential(*list(model.children())[:-1])

model.eval()
```

```python
# Transformaciones

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
# Función para extraer embedding

def extract_embedding(img_path):
    img = Image.open(img_path).convert("RGB")
    img = transform(img).unsqueeze(0)

    with torch.no_grad():
        emb = model(img)

    return emb.squeeze().numpy()
```

```python
# Recorrer datasets (cogiendo 1 imagen por media y extrayendo embedding)

rows = []

dataset_root = "database"

for dataset_name in os.listdir(dataset_root):

    dataset_path = os.path.join(dataset_root, dataset_name)

    if not os.path.isdir(dataset_path):
        continue

    for media_name in os.listdir(dataset_path):

        media_path = os.path.join(dataset_path, media_name)
        images_path = os.path.join(media_path, "images")

        if not os.path.exists(images_path):
            continue

        image_files = [
            f for f in os.listdir(images_path)
            if f.lower().endswith((".jpg", ".jpeg", ".png"))
        ]

        if len(image_files) == 0:
            continue

        # Elegimos imagen aleatoria
        img_file = random.choice(image_files)

        img_path = os.path.join(images_path, img_file)

        try:
            embedding = extract_embedding(img_path)

            row = {
                "dataset_id": dataset_name,
                "media_id": media_name,
            }

            for i, val in enumerate(embedding):
                row[f"f_{i}"] = val

            rows.append(row)

            print(f"Processed {dataset_name}/{media_name}")

        except Exception as e:
            print(f"Error with {img_path}: {e}")
```

```python
# Crear dataframe

emb_df = pd.DataFrame(rows)

emb_df.to_csv("output/deep_embeddings.csv", index=False)

emb_df.head()
```

```python
# PCA 2D

features = emb_df.filter(regex="f_").values

pca = PCA(n_components=2)

X_pca = pca.fit_transform(features)

plot_df = pd.DataFrame({
    "x": X_pca[:,0],
    "y": X_pca[:,1],
    "dataset": emb_df["dataset_id"]
})
```

```python
# Visualización

plt.figure(figsize=(10,8))

for dataset in plot_df["dataset"].unique():

    subset = plot_df[plot_df["dataset"] == dataset]

    plt.scatter(
        subset["x"],
        subset["y"],
        label=dataset,
        alpha=0.7
    )

plt.legend()
plt.title("PCA of Deep Embeddings")

plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(
    n_components=2,
    perplexity=15,
    random_state=42
)

X_tsne = tsne.fit_transform(features)

plot_df = pd.DataFrame({
    "x": X_tsne[:,0],
    "y": X_tsne[:,1],
    "dataset": emb_df["dataset_id"]
})

plt.figure(figsize=(12,10))

for dataset in plot_df["dataset"].unique():

    subset = plot_df[plot_df["dataset"] == dataset]

    plt.scatter(
        subset["x"],
        subset["y"],
        label=dataset,
        alpha=0.7
    )

plt.legend()
plt.title("t-SNE of Deep Embeddings")

plt.show()
```

```python
# ============================================================
# EXTRACT DEEP EMBEDDINGS + PCA + t-SNE VISUALIZATION
# ============================================================

import os
import random
from pathlib import Path

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

import torch
import torchvision.models as models
import torchvision.transforms as transforms

# ============================================================
# CONFIG
# ============================================================

DATASETS = {
    "Turtle": "database/YoutubeTurtles",
    "Shark": "database/RoboflowSharks",
    "Whale": "database/YoutubeWhales"
}

MAX_IMAGES_PER_CLASS = 150
IMAGE_SIZE = 224
RANDOM_SEED = 42

random.seed(RANDOM_SEED)

# ============================================================
# LOAD PRETRAINED RESNET18
# ============================================================

model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# Remove classification layer
model = torch.nn.Sequential(*list(model.children())[:-1])

model.eval()

# ============================================================
# IMAGE TRANSFORMS
# ============================================================

transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# ============================================================
# LOAD IMAGE PATHS
# ============================================================

samples = []

for class_name, dataset_path in DATASETS.items():

    image_paths = list(Path(dataset_path).rglob("*.jpg"))
    image_paths += list(Path(dataset_path).rglob("*.png"))
    image_paths += list(Path(dataset_path).rglob("*.jpeg"))

    random.shuffle(image_paths)

    image_paths = image_paths[:MAX_IMAGES_PER_CLASS]

    for path in image_paths:

        samples.append({
            "class": class_name,
            "path": path
        })

print(f"\nLoaded {len(samples)} images")

# ============================================================
# EXTRACT EMBEDDINGS
# ============================================================

embeddings = []
labels = []

for sample in tqdm(samples):

    image = cv2.imread(str(sample["path"]))

    if image is None:
        continue

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    tensor = transform(image).unsqueeze(0)

    with torch.no_grad():

        embedding = model(tensor)

    embedding = embedding.squeeze().numpy()

    embeddings.append(embedding)
    labels.append(sample["class"])

embeddings = np.array(embeddings)

print("\nEmbeddings shape:", embeddings.shape)

# ============================================================
# NORMALIZE
# ============================================================

scaler = StandardScaler()

embeddings_scaled = scaler.fit_transform(embeddings)

# ============================================================
# PCA
# ============================================================

pca = PCA(n_components=2)

pca_result = pca.fit_transform(embeddings_scaled)

pca_df = pd.DataFrame({
    "PCA1": pca_result[:, 0],
    "PCA2": pca_result[:, 1],
    "Class": labels
})

# ============================================================
# PCA PLOT
# ============================================================

plt.figure(figsize=(10, 8))

for class_name in pca_df["Class"].unique():

    subset = pca_df[pca_df["Class"] == class_name]

    plt.scatter(
        subset["PCA1"],
        subset["PCA2"],
        label=class_name,
        alpha=0.7
    )

plt.title("PCA sobre embeddings profundos")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
plt.grid(True)

plt.show()

# ============================================================
# t-SNE
# ============================================================

tsne = TSNE(
    n_components=2,
    perplexity=30,
    random_state=RANDOM_SEED
)

tsne_result = tsne.fit_transform(embeddings_scaled)

tsne_df = pd.DataFrame({
    "TSNE1": tsne_result[:, 0],
    "TSNE2": tsne_result[:, 1],
    "Class": labels
})

# ============================================================
# t-SNE PLOT
# ============================================================

plt.figure(figsize=(10, 8))

for class_name in tsne_df["Class"].unique():

    subset = tsne_df[tsne_df["Class"] == class_name]

    plt.scatter(
        subset["TSNE1"],
        subset["TSNE2"],
        label=class_name,
        alpha=0.7
    )

plt.title("t-SNE sobre embeddings profundos")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.legend()
plt.grid(True)

plt.show()
```
